Gradient Descent on 2D surface:

```python
import numpy as np
import matplotlib.pyplot as plt

# Function: y = x^2
def y_function(x):
    return x ** 2

# Derivative: dy/dx = 2x
def y_derivative(x):
    return 2 * x

# Define x values for visualization
x = np.linspace(-10, 10, 400)
y = y_function(x)

# Gradient Descent Parameters
learning_rate = 0.1
iterations = 50
current_x = 8  # Starting point

# Store descent path
x_history = [current_x]
y_history = [y_function(current_x)]

# Perform gradient descent
for _ in range(iterations):
    current_x -= learning_rate * y_derivative(current_x)
    x_history.append(current_x)
    y_history.append(y_function(current_x))

# Plot function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = x^2", color="blue")
plt.scatter(x_history, y_history, color="red", s=15, label="Gradient Descent Path")
plt.scatter(x_history[-1], y_history[-1], color="green", s=100, label="Final Point")  # Final point

# Labels and Title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent on y = x^2")
plt.legend()
plt.show()
```

Gradient Descent 3D Surface:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function: z = x^2 + y^2
def z_function(x, y):
    return x ** 2 + y ** 2

# Partial derivatives
def dz_dx(x, y):
    return 2 * x

def dz_dy(x, y):
    return 2 * y

# Create meshgrid for function surface
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

# Gradient Descent Parameters
learning_rate = 0.1
iterations = 50
current_x, current_y = 8, 8  # Start far from the minimum

# Store descent path
x_history, y_history, z_history = [current_x], [current_y], [z_function(current_x, current_y)]

# Perform gradient descent
for _ in range(iterations):
    grad_x = dz_dx(current_x, current_y)
    grad_y = dz_dy(current_x, current_y)
    
    current_x -= learning_rate * grad_x
    current_y -= learning_rate * grad_y
    
    x_history.append(current_x)
    y_history.append(current_y)
    z_history.append(z_function(current_x, current_y))

# 3D Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the function surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Plot the gradient descent path
ax.scatter(x_history, y_history, z_history, color="red", s=20, label="Gradient Descent Path")
ax.scatter(x_history[-1], y_history[-1], z_history[-1], color="green", s=100, label="Final Point")

# Labels and title
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z = x^2 + y^2")
ax.set_title("Gradient Descent on 3D Surface")
ax.legend()

plt.show()
```
